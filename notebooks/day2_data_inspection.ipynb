# Day 2 - Data Inspection

- Loaded IPL dataset
- Checked missing values
- Filtered seasons 2021–2023

```python
import pandas as pd

deliveries = pd.read_csv("../data/raw/deliveries.csv")
deliveries.info()
deliveries.isnull().sum()
```

```python
matches = pd.read_csv("../data/raw/matches.csv")
matches.head()
```

```python
matches.columns
```

```python
matches_2021_2023 = matches[
    matches["season"].isin(["2021", "2022", "2023"])
]

matches_2021_2023.head()
```

```python
match_ids = matches_2021_2023["id"]
```

```python
deliveries = deliveries[
    deliveries["match_id"].isin(match_ids)
]
```

```python
deliveries.shape
```

```python
deliveries.to_csv(
    "../data/processed/deliveries_2021_2023.csv",
    index=False
)
```
